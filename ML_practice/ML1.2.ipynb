# Практические задания

### 1. Построение от каждого атрибута

```python
for atribute in range(1, 5):
    binary = LinearRegression()
    binary.fit(x[[atribute]], y)
    print(f"Построение модели от столбца №{atribute}")
    print(f"Коэффициент = {binary.coef_}")
    print(f"Score модели = {binary.score(x[[atribute]], y)}")
    
    plt.scatter(x[atribute], y)
    xx = np.linspace(x[atribute].min(), x[atribute].max(), 100).reshape((-1, 1))
    plt.plot(xx, binary.predict(xx), c='r')
    plt.title(f"График от {atribute} атрибута")
    plt.show()

    history_score.append(binary.score(x[[atribute]], y))
    name_regression.append(f"LinearRegression for atribute {atribute}")
    
    print('-' * 6)
```

### 2. Произвольное количество столбцов

```python
for atribute in x.columns:
    binary = LinearRegression()
    binary.fit(x[[atribute]], y)
    print(f"Построение модели от столбца №{atribute}")
    print(f"Коэффициент = {binary.coef_}")
    print(f"Score модели = {binary.score(x[[atribute]], y)}")
    
    plt.scatter(x[atribute], y)
    xx = np.linspace(x[atribute].min(), x[atribute].max(), 100).reshape((-1, 1))
    plt.plot(xx, binary.predict(xx), c='r')
    plt.title(f"График от {atribute} атрибута")
    plt.show()
    
    print('-' * 6)
```

### 3. Полиномиальная регрессия

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
for attribute in x.columns:
    print(f"Построение полиномиальных моделей от столбца №{attribute}")
    
    for polinom in [2, 3, 10]:
        poly = PolynomialFeatures(degree=polinom, include_bias=False)
        x_poly = poly.fit_transform(x[[attribute]])
        
        model = LinearRegression()
        model.fit(x_poly, y)
        
        print(f"Степень полинома {polinom}:")
        print(f"  Коэффициенты = {model.coef_}")
        print(f"  Score модели (R^2) = {model.score(x_poly, y)}")

        plt.scatter(x[attribute], y)
        
        x_sorted = np.sort(x[attribute])
        x_plot = x_sorted.reshape(-1, 1)
        x_plot_poly = poly.transform(x_plot)
        y_plot = model.predict(x_plot_poly)
        
        plt.plot(x_sorted, y_plot, label=f'степень {polinom}')
        plt.title(f"Полиномиальная регрессия для атрибута №{attribute}")
        plt.legend()
        plt.show()

        history_score.append(model.score(x_poly, y))
        name_regression.append(f"PolynomialFeatures for atribute {attribute} and polinom {polinom}")
    
    print('-' * 60)
```

```python
sheet_scores = pd.DataFrame({
    'name_regression': name_regression,
    'history_score': history_score
})
```

```python
sheet_scores
```

```python
sheet_scores.sort_values('history_score', ascending=False)
```

Анализ качества моделей

1. **Лучшая модель:** Линейная регрессия для атрибута 1 (R² = 0.992) - практически идеальное качество

2. **Переобученные модели:**
   - Полином 10 для атрибута 1 (R² = 0.773) - значительно ниже, чем у линейной модели (0.992). Высокая степень полинома привела к потере обобщающей способности
   - Полином 10 для атрибута 2,3,4,0 - качество не улучшилось по сравнению с линейными моделями

3. **Недообученные модели:**
   - Все модели для атрибутов 2,3,4,0 (R² < 0.2) - слабая зависимость между этими признаками и целевой переменной
   - Полиномиальные модели для этих атрибутов не улучшили качество, что указывает на отсутствие нелинейной зависимости

```python
correlations = x.corrwith(y.iloc[:, 0])

correlations.plot(kind='bar')
plt.axhline(y=0, linewidth=0.5)
plt.xlabel('Атрибуты')
plt.ylabel('Коэффициент корреляции')
plt.title('Корреляция атрибутов с целевой переменной')
plt.show()
```

### 5. Другие метрики качества 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
results = []

for attribute in x.columns:
    linear = LinearRegression()
    linear.fit(x[[attribute]], y)
    y_pred_linear = linear.predict(x[[attribute]])
    
    mse = mean_squared_error(y, y_pred_linear)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y, y_pred_linear)
    r2 = linear.score(x[[attribute]], y)
    
    results.append({
        'Модель': f'Линейная (атр.{attribute})',
        'R2': r2,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae
    })
    
    for degree in [2, 3, 10]:
        poly = PolynomialFeatures(degree=degree, include_bias=False)
        x_poly = poly.fit_transform(x[[attribute]])
        
        model = LinearRegression()
        model.fit(x_poly, y)
        y_pred = model.predict(x_poly)
        
        mse = mean_squared_error(y, y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y, y_pred)
        r2 = model.score(x_poly, y)
        
        results.append({
            'Модель': f'Полином {degree} (атр.{attribute})',
            'R2': r2,
            'MSE': mse,
            'RMSE': rmse,
            'MAE': mae
        })

results_df = pd.DataFrame(results)
print(results_df.sort_values('R2', ascending=False))

multiple = LinearRegression()
multiple.fit(x, y)
y_pred_mult = multiple.predict(x)

print("\nМножественная регрессия:")
print(f"R2 = {multiple.score(x, y):.6f}")
print(f"MSE = {mean_squared_error(y, y_pred_mult):.6f}")
print(f"RMSE = {np.sqrt(mean_squared_error(y, y_pred_mult)):.6f}")
print(f"MAE = {mean_absolute_error(y, y_pred_mult):.6f}")
```
