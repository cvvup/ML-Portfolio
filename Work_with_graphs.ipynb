```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
#plotli express, но лучшим считается plt
```

```python
students_performance = pd.read_csv("https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv")
```

```python
students_performance
```

```python
students_performance = students_performance.rename(columns = {'parental level of education' : 'parental_level_of_education', 'test preparation course' : 'test_preparation_course', 'math score' : 'math_score', 'reading score' : 'reading_score', 'writing score' : 'writing_score'})
```

```python
students_performance.math_score.hist()
```

```python
students_performance.plot.scatter(x = 'math_score', y = 'reading_score')
```

```python
ax = sns.lmplot (x = 'math_score', y = 'reading_score', hue = 'gender', data = students_performance)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
